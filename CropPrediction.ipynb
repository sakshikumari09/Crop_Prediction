```python
X = df[['N', 'P', 'temperature', 'humidity', 'ph', 'rainfall']]
Y = df[['crop_num']]
X.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test,Y_train, Y_test = train_test_split(X,Y, test_size=0.2, random_state =0)
```

```python
print(X_train.shape)
print(Y_train.shape)
```

```python
from sklearn.preprocessing import MinMaxScaler
ms = MinMaxScaler()

X_train = ms.fit_transform(X_train)
X_test = ms.transform(X_test)
```

```python
X_train
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()


sc.fit(X_train)
X_train = sc.transform(X_train)
X_test = sc.transform(X_test)
```

```python
#models 
#since it is a classification problem
#we shall run  KNN, Naive baiyes, logistic regression , random forest, decision tree
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
```

```python
# create instances of all models
models = {
    'Logistic Regression': LogisticRegression(),
    'Naive Bayes': GaussianNB(),
    
    'K-Nearest Neighbors': KNeighborsClassifier(),
    'Decision Tree': DecisionTreeClassifier(),
    'Random Forest': RandomForestClassifier(),
    
}
```

```python
from sklearn.metrics import accuracy_score
for name, md in models.items():
    md.fit(X_train,Y_train)
    ypred = md.predict(X_test)
    
    print(f"{name}  with accuracy : {accuracy_score(Y_test,ypred)}")
```

```python
#we clearly see random forest works best.
rfc = RandomForestClassifier()
rfc.fit(X_train,Y_train)
ypred = rfc.predict(X_test)
accuracy_score(Y_test,ypred)
```

```python
knnq =  KNeighborsClassifier()
knnq.fit(X_train,Y_train)
ypred = knnq.predict(X_test)
accuracy_score(Y_test,ypred)
```

```python
def recommendation(N,P,temperature,humidity,ph,rainfal):
    features = np.array([[N,P,temperature,humidity,ph,rainfal]])
    transformed_features = ms.fit_transform(features)
    transformed_features = sc.fit_transform(transformed_features)
    prediction = knnq.predict(transformed_features).reshape(1,-1)
    
    return prediction[0] 
    
```

```python
N = 99
P = 15
temp = 28
hum = 56.6
ph = 6.08
rf= 127.9
predict = recommendation(N,P,temp,hum,ph,rf)


crop_dict = {1: "Rice", 2: "Maize", 3: "Jute", 4: "Cotton", 5: "Coconut", 6: "Papaya", 7: "Orange",
                 8: "Apple", 9: "Muskmelon", 10: "Watermelon", 11: "Grapes", 12: "Mango", 13: "Banana",
                 14: "Pomegranate", 15: "Lentil", 16: "Blackgram", 17: "Mungbean", 18: "Mothbeans",
                 19: "Pigeonpeas", 20: "Kidneybeans", 21: "Chickpea", 22: "Coffee"}

if predict[0] in crop_dict:
    df= crop_dict[predict[0]]
    print("{} is a best crop to be cultivated ".format(df))
else:
    print("Sorry are not able to recommend a proper crop for this environment")
```
